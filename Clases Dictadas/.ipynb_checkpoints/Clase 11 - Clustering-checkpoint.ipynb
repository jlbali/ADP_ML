```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans, DBSCAN
% matplotlib inline
```

```python
# "cluster 1"
x = np.random.uniform(low=0.0, high=10.0,size=100).reshape(100,1)
y = np.random.uniform(low=0.0, high=10.0,size=100).reshape(100,1)
X1 = np.concatenate([x,y], axis=1)
# "cluster 2"
x = np.random.uniform(low=15.0, high=25.0,size=100).reshape(100,1)
y = np.random.uniform(low=0.0, high=10.0,size=100).reshape(100,1)
X2 = np.concatenate([x,y], axis=1)
# Juntar todo.
X = np.concatenate([X1,X2], axis=0)
# Graficar
plt.scatter(X[:,0], X[:,1])

```

```python
clusterizador = KMeans(n_clusters=2)
```

```python
y = clusterizador.fit_predict(X)
y
```

```python
plt.scatter(X[y==0, 0], X[y==0, 1])
plt.scatter(X[y==1, 0], X[y==1, 1])
```

```python
# "cluster 1"
x = np.random.uniform(low=0.0, high=10.0,size=100).reshape(100,1)
y = np.random.uniform(low=0.0, high=1000.0,size=100).reshape(100,1)
X1 = np.concatenate([x,y], axis=1)
# "cluster 2"
x = np.random.uniform(low=15.0, high=25.0,size=100).reshape(100,1)
y = np.random.uniform(low=0.0, high=1000.0,size=100).reshape(100,1)
X2 = np.concatenate([x,y], axis=1)
# Juntar todo.
X = np.concatenate([X1,X2], axis=0)
# Graficar
plt.scatter(X[:,0], X[:,1])
```

```python
clusterizador = KMeans(n_clusters=4)
y = clusterizador.fit_predict(X)
plt.scatter(X[y==0, 0], X[y==0, 1])
plt.scatter(X[y==1, 0], X[y==1, 1])
plt.scatter(X[y==2, 0], X[y==2, 1])
plt.scatter(X[y==3, 0], X[y==3, 1])
```

```python
# "cluster 1"
x = np.random.uniform(low=0.0, high=10.0,size=100).reshape(100,1)
y = np.random.uniform(low=0.0, high=10.0,size=100).reshape(100,1)
X1 = np.concatenate([x,y], axis=1)
# "cluster 2"
x = np.random.uniform(low=15.0, high=25.0,size=100).reshape(100,1)
y = np.random.uniform(low=0.0, high=10.0,size=100).reshape(100,1)
X2 = np.concatenate([x,y], axis=1)
# Juntar todo.
X = np.concatenate([X1,X2], axis=0)
# Graficar
plt.scatter(X[:,0], X[:,1])
```

```python
inertias = []
for i in range(1,10):
    clusterizador = KMeans(n_clusters=i)
    y = clusterizador.fit_predict(X)
    inertias.append(clusterizador.inertia_)
plt.plot(range(1,10), inertias)
    
```

```python
# "cluster 1"
x = np.random.uniform(low=0.0, high=10.0,size=100).reshape(100,1)
y = np.random.uniform(low=0.0, high=10.0,size=100).reshape(100,1)
X1 = np.concatenate([x,y], axis=1)
# "cluster 2"
x = np.random.uniform(low=15.0, high=25.0,size=100).reshape(100,1)
y = np.random.uniform(low=0.0, high=10.0,size=100).reshape(100,1)
X2 = np.concatenate([x,y], axis=1)
# "cluster 3"
x = np.random.uniform(low=0.0, high=10.0,size=100).reshape(100,1)
y = np.random.uniform(low=15.0, high=25.0,size=100).reshape(100,1)
X3 = np.concatenate([x,y], axis=1)
# "cluster 4"
x = np.random.uniform(low=15.0, high=25.0,size=100).reshape(100,1)
y = np.random.uniform(low=15.0, high=25.0,size=100).reshape(100,1)
X4 = np.concatenate([x,y], axis=1)
# Juntar todo.
X = np.concatenate([X1,X2,X3,X4], axis=0)
# Graficar
plt.scatter(X[:,0], X[:,1])
```

```python
inertias = []
for i in range(1,10):
    clusterizador = KMeans(n_clusters=i)
    y = clusterizador.fit_predict(X)
    inertias.append(clusterizador.inertia_)
plt.plot(range(1,10), inertias)
```
