```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
from sklearn.linear_model import LinearRegression, Lasso
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, classification_report
% matplotlib inline
```

```python
# Dataset artificial "planeta vs orbita"
```

```python
# Planeta (clase 0)
n0 = 100
r = np.random.uniform(low=0.0, high=5.0, size=n0)
angulos = np.random.uniform(low=0.0, high=2*np.pi,size=n0)
x1 = (r*np.cos(angulos)).reshape(n0,1)
x2 = (r*np.sin(angulos)).reshape(n0,1)
plt.scatter(x1,x2)
X0 = np.concatenate([x1,x2], axis=1)
Y0 = np.zeros((n0,1))
```

```python
# Orbita (clase 1)
n1 = 200
r = np.random.uniform(low=4.0, high=8.0, size=n1)
angulos = np.random.uniform(low=0.0, high=2*np.pi,size=n1)
x1 = (r*np.cos(angulos)).reshape(n1,1)
x2 = (r*np.sin(angulos)).reshape(n1,1)
plt.scatter(x1,x2)
X1 = np.concatenate([x1,x2], axis=1)
Y1 = np.ones((n1,1))
```

```python
# Plot conjunto
plt.scatter(X0[:,0], X0[:,1])
plt.scatter(X1[:,0], X1[:,1])
plt.grid()
```

```python
# Juntar todo...
X = np.concatenate([X0, X1], axis=0)
Y = np.concatenate([Y0, Y1], axis=0)
Y = Y.reshape(n0+n1)
```

```python
# Vamos a entrenar con todo el dataset completo
# y vamos a validar con todo el dataset completo.
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X,Y)
```

```python
Y_pred = knn.predict(X)
```

```python
misma_clase =  (Y == Y_pred)
aciertos = misma_clase.sum()
tasa = aciertos / len(misma_clase)
```

```python
tasa
```

```python
# Separacion de un conjunto de validacion.
n_total = n0 + n1
total = range(n_total)
index_train = np.random.choice(n_total, size=int(0.8*n_total), replace=False)
index_test = np.setdiff1d(total, index_train)
```

```python
X_train = X[index_train,:]
Y_train = Y[index_train]
X_test = X[index_test, :]
Y_test = Y[index_test]
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, Y_train)
Y_pred = knn.predict(X_test)
```

```python
tasa_aciertos = (Y_test == Y_pred).sum() / len(Y_test)
tasa_aciertos
```

```python
confusion_matrix(Y_test, Y_pred)
```

```python
# (0,0) -> TN
# (1,0) -> FN
# (0,1) -> FP
# (1,1) -> TP
```

```python
print(classification_report(Y_test, Y_pred))
```

```python
tasas = []
for k in range(1,240):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, Y_train)
    Y_pred = knn.predict(X_test)
    tasa = (Y_pred == Y_test).sum() /len(Y_pred)
    tasas.append(tasa)
plt.plot(range(1,240), tasas)
```

```python
np.argmax(tasas)
```
